```python
def allocation_statistics(folder_name):
    Cov = pd.read_csv(folder_name + "server_allocation.csv", sep=',')[a:b]
    Cov.columns = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    columns_titles = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    df = Cov.divide(Cov.sum(axis=1), axis=0) * 100
    df = df.reindex(columns=columns_titles)    
    color = {'BS':'purple', 'VS1':'mediumblue', 'VS2':'cyan', 'VS3':'g', 
             'VS4':'yellow', 'VS5':'orange', 'Cloud':'red'}
    df.plot.area(legend = False, color = color)
    plt.xticks([*range(60, 81, 4)], [*range(0, 21, 4)])
    plt.xlabel("Time slot")
    plt.ylabel("Percentage of task offloading (%)")
    plt.savefig("../../pics/algs/allo_"+ folder_name[:-3] +".pdf")
    
def extra_statistics(folder_name):
    Cov = -pd.read_csv(folder_name + "extra_allocation.csv", sep=',')[a:b]
    Cov.columns = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    columns_titles = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    df = Cov * 1000
    df = df.reindex(columns=columns_titles)    
    color = {'BS':'purple', 'VS1':'mediumblue', 'VS2':'cyan', 'VS3':'g', 
             'VS4':'yellow', 'VS5':'orange', 'Cloud':'red'}
    df.plot.area(legend = False, color = color)
    plt.xticks([*range(60, 81, 4)], [*range(0, 21, 4)])
    plt.xlabel("Time slot")
    plt.ylabel("Tolerance time per task (ms)")
    plt.savefig("../../pics/algs/extra_"+ folder_name[:-3] +".pdf")
    
def delay_statistics(folder_name):
    Cov = pd.read_csv(folder_name + "delay_allocation.csv", sep=',')[a:b]
    Cov.columns = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    columns_titles = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5", "Cloud"]
    df = Cov * 1000
    df = df.reindex(columns=columns_titles)    
    color = {'BS':'purple', 'VS1':'mediumblue', 'VS2':'cyan', 'VS3':'g', 
             'VS4':'yellow', 'VS5':'orange', 'Cloud':'red'}
    df.plot.area(legend = False, color = color)
    plt.xticks([*range(60, 81, 4)], [*range(0, 21, 4)])
    plt.xlabel("Time slot")
    plt.ylabel("Delay time per task (ms)")
    plt.savefig("../../pics/algs/delay_"+ folder_name[:-3] +".pdf")
#allocation_statistics(folder_name5 + str(2) + '/')
```
